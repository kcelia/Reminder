# Markdown:

Markdown est un moyen de styliser du texte sur le Web.

L'extension est: *.md* ou *.markdown*.

# Headers:

Vous pouvez utiliser un jusqu'à six `######` pour différentes tailles de titre.


# 1.TEXT

**Code:**
```bash
# 1.TEXT
```

2.TEXT
=============================================

**Code:**
```bash
2.TEXT
=============================================
```

3.TEXT
_____________________________________________

**Code:**
```bash
3.TEXT
_____________________________________________
```

4.TEXT
---------------------------------------------

**Code:**
```md
4.TEXT
---------------------------------------------
```

## Citations:

Commencer la ligne avec `>` pour faire une citation:


> Citation 1
> - Sub citation 1

> Citation 2
>> Sub citation 2.1
>>
>> Sub citation 2.2
>>> * Puce
>>> * Autre puce

## List:

1. Liste detirets:

	- [x] Finish my changes
	- [ ] Push my commits to GitHub
	- [ ] Open a pull request
    
1. Liste à puce:

	* tiret1 % '*' avec saut de ligne
	- tiret2 % '-' sans de saut de ligne
	- tiret3 % avec 1 tabulation
		- tiret4 % avec 2 tabulations
	> tiret5
	+ tiret 6 % '+' avec saut de ligne

    Remarque : '-' et '*' produisent la même enumération.

1. Numérotés:

	1. Item 1
	1. Item 2
	1. Item 3
		1. Item 3a
		1. Item 3b
