## <span style="color:blue">1. Injective</span>

- <span style="color:red">$f$ est **injective** si $\forall x, x_1 \in E \;\;\; (f(x) = f(x_1) \implies x = x_1)$</span>

- <span style="color:red">$\forall y \in F$ a au plus 1 antécédant (éventuellement aucun)</span>

- Une fonction n'est pas injective $ssi$ $\exists x, x_1 \in E \;\;\; (f(x) = f(x_1) \; et \; x \neq x_1)$

![](/application_injective.png)



<span style="color:green"><u>Exemple 1:</u></span>

$$f: N \to N$$

$$\;\;\;\; n \to n + 1$$

Soient $n, n_1 \in N \;\; lq \;\;f(n) = f(n_1)$


$\implies n + 1 = n_1 + 1$

$\implies n = n_1$

$\implies$ f est injective

<span style="color:green"><u>Exemple 2:</u></span>

$$f: R_2 \implies R_2$$

$$\;\; (x, y) \implies (x + y, x - y)$$

Soient $ (x, y), \; (x_1, y_1) \in R_2$ tq $f(x, y) = f(x_1, y_1)$

$\implies (x + y, x - y) = (x_1 + y_1, x_1 - y_1)$

$\implies \begin{cases} x + y =  x_1 + y_1 & \text{} \\ x - y = x_1 - y_1 & \text{} \end{cases} $

$\implies \begin{cases} 2x =  2x_1 & \text{} \\ 2y = 2y_1 & \text{} \end{cases} $

$\implies (x, y) = (x_1, y_1)$

$\implies$ f est injective

<span style="color:green"><u>Exemple 3:</u></span>

$$f: R_2 \to R$$

$$x \to x^2$$

Soient $x, x_1 \in \mathbb{R} \;\; lq \;\; f(x) = f(x_1)$

$\implies x^2 = x_1^2 $

$\implies x = x_1 \;\; ou \;\; x = -x_1$

$ f(1) = 1 \;\; et f(-1) = 1$

$\implies$ f n'est pas injective

<span style="color:green"><u>Exemple 4:</u></span>

$$f: R_2 \to R$$

$$\;\;x \to {x \over x^2 + 1}$$

Soient $x, x_1 \in \mathbb{R} \;\; lq \;\; f(x) = f(x_1)$

$\implies {x \over x^2 + 1} = {x_1 \over x_1^2 + 1}$

$\implies x * (x_1^2 + 1) = x_1 * (x^2 + 1) $

$\implies xx_1^2 + x = x_1x^2 + x_1 $

$\implies xx_1(x_1 - x) +x -x_1 = 0 $

$\implies (x_1 - x)( xx_1 - 1) = 0 $

$\implies x_1 = x \;\; ou \;\; xx_1 = 1 $

$\implies x_1 = x \;\; ou \;\; xx_1 = 1 $

$Pour x = 1, \; xx_1 = 1 \implies x_1 = 1/2$

$ f(2) = 2/5, \; f(1/2) = 2/5 \implies f(2) = f(1/2) \;\; 2 != 2/5$

$\implies$ f n'est pas injective

## <span style="color:blue">2. Surjective</span>

- <span style="color:red">$f$ est **surjective** si $\forall y \in F, \;\;\; \exists x \in E \;\;\; (y = f(x))$</span>

- F est **surjective** ssi $f(E) = F$

- <span style="color:red">$\forall y \in F$ a au moins 1 antécédant</span>

- E.g.: Fonction surjective: fonction avec plusieurs minima ou maxima locaux 

- Une fonction n'est pas injective $ssi \; \; \exists y \in F, \; \forall x \in E \; y \neq f(x)$

![](application_surjective2.png)



<span style="color:green"><u>Exemple 1:</u></span>

$$f: \mathbb{N} \to \mathbb{N}$$

$$n \to n + 1$$

$y = 0$ n'a pas d'an antécédant dans \mathbb{N}$

$\implies f$ n'est pas surjective

<span style="color:green"><u>Exemple 2:</u></span>

$$f: \mathbb{Z} \to \mathbb{Z}$$

$$n \to n + 1$$

soit $m \in \mathbb{Z}$, on pose $n = m - 1 $

$g(n) m - 1 + 1 = m $

$\forall n \in \mathbb{Z}$ on a un antécédant $m \in \mathbb{Z}$

$\implies f$ est surjective

<span style="color:green"><u>Exemple 3:</u></span>

$$f: R_2 \to R_2$$

$$(x, y) \to (x + y, x - y)$$

Soit $ (X, Y) \in \mathbb{R_2}$, on cherche $(x, y) \in \mathbb{R_2}$, tq $f(x, y) = (X, Y)$

$\implies (x + y, x - y) = (X, Y)$

$\implies \begin{cases} x + y = X & \text{} \\ x - y = Y & \text{} \end{cases} $

$\implies \begin{cases} 2x =  X + Y & \text{} \\ 2y = X - Y & \text{} \end{cases} $

$\implies (x, y) = ({X + Y \over 2}, {X - Y \over 2})$

$(x, y) = ({X + Y \over 2})$ est un antécédent de $(X, Y)$

$\implies$ f est surjective 

## <span style="color:blue">3. Bijective</span>

$f$ est bijective si elle est injective et surjective.

- Tout élément de F a un unique antécédent par f

$\forall y \in F, \; \exists !x \in E, \; tq \; (y = f(x))$

**Proposition**
- L'application $f: \mathbb{E} \implies \mathbb{F}$ est bijective s'il existe: $\mathbb{F} \implies \mathbb{E}$ telle que $f o g = id_F \; et \; g o f = id_E$. 

    - $f o g =id_F $ signifie $ \forall y \in F \;\; f(g(y)) = y$
    - $g o f =id_E $ signifie $ \forall x \in E \;\; g(f(x)) = x$

- Si $f$ est bijective alors $g$ est unique et est bijective, c'est la **bijection réciproque** de f et est notée $f^{-1}$. De plus $(f^{-1})^{-1} = f$

- Soient $f: \mathbb{E} \implies \mathbb{F}$ et $g: \mathbb{F} \implies \mathbb{E}$ des applications bijectives. L'application $g o f$ est bijective et $(g o f)^{-1} = g^{-1} o f^{-1}$

![](application_bijective_exp_ln.png)



<span style="color:green"><u>Exemple 1:</u></span>

$$f: [1, +\infty[ \; \to [0, +\infty[ $$

$$ f(x) \; \to  x^2 - 1$$

*Méthode: Etude de $f$*
- $f$ est une fonction polynomiale, $f$ est donc continue
- $f^\prime(x) = 2x$, $f$ est donc strictement croissante 
- $f(1) = 0$ et $\lim\limits_{x \to +\infty} {f} = +\infty$

$\implies f$ est bijective 

<span style="color:blue">Structure Algébrique</span>
===

> Structuer un ensemble par des opérateurs algébriques.

- Soit $E$ un ensemble non vide, qui peut être: naturels, réels, des complexes, fonctions, polynomes, droites dans le plan...

- Une loi de composition interne (LCI) est une application qui à partir de 2 éléments de $E$, associe un aute élément de $E$. C'est une opération binaire par laquelle $E$ est stable. 

    <span style="color:green"><u>Exemple:</u></span> $(N, +)$ L'ensemble des entiers natuel et la loi d'addition usuelle:

    $2\in \mathbb{N}$, $2\in \mathbb{N}$, $2 + 3 = 5 \in \mathbb{N}$

**Quelques définitions**:

- La commutativité $\forall x, y \in E, \; x*y = y*x$

- L'associativité $\forall x, y, z \in E, \; (x*y)* z = x*(y*z)$

- Si la LCI est l'addition: 
    - L'élément neutre, noté $e$, vérifie : $\forall x \in E, \exists e \in E, \; e = 0, \; x + e = e + x = x$
    - L'élément symetrique (opposé) $\forall x \in E, \; \exists -x \in E, \; x + (-x)  = (-x) + x = e$

- Si la LCI est la multiplication: 
    - L'élément neutre, noté $e$, vérifie : $\forall x \in, \exists e \in E, \; e = 1, \; x*e = e*x = x$
    - L'élément symétrique (inverse) $\forall x \in E^*, \; \exists {1 \over x} \in E, \; x + {1 \over x}  = {1 \over x} + x = e$  


<span style="color:red"><u>$(G, \star)$ est un group ssi il possède:</u></span>

- $\star$ est une LCI sur $G$
- Il existe un élément neutre $e \in G$ pour la LCI $\star$. Autrement dit, si on compose un élément $x$ de $G$ avec l'élement neutre $e$, on garde l'élément $x$.
$\forall x \in G, \; \exists e, \;  x \star e = e  \star x = x$ 

⚠️ Si $(G, \star) $ est un groupe quelconque, on n'a pas: $\forall x, y \in G, \; x \star y = y \star x$

⚠️ Si dans $(G, \star)$ la LCI $\star$ est commutatif, alors $(G, \star)$ est un groupe _abelien_ 

- Tout élément de $G$, admet un symétrique dans $G$ pour la LCI $\star$. Autrement dit, pour tout élément de $G$, il existe un autre élément de $G$, tel que quand on les combine avec la LCI, on retombe sur l'élément neutre.
$\forall x \in G, \; \exists y \in G, \; x \star y = y \star x = e$

- La LCI $\star$ est associative.
$\forall (a, b, c) \in G^3, \;  a \star (b \star c)  = (a \star b) \star c$


Exemple: 

$(Z, +)$ est un groupe abelien 



<span style="color:red"><u>$(A, +, *)$ est un anneau ssi:</u></span>

- $(A, +)$ est un groupe abelien
- $*$ est une LCI sur A
- $*$ est associative et distributive sur $+\;$: $\forall (a, b, c) \in G^3, \;  a * (b + c)  = a * b + a * c$
- $*$ admet un élément neutre dans A, noté $1_A$ ($0_A$ pour +)


<span style="color:red"><u>$(K, +, *)$ est un corps ssi:</u></span>

- $(K, +, *)$ est un anneau abelien
- $\forall x \in K, \;  x \neq 0_A, \; \exists x^\prime \in K, \;  x * x^\prime = x^\prime * x = 1_A$

<span style="color:blue">Espace vectoriel</span>
===
Un espace vectoriel est un ensemble munie de 2 lois:
- Addition 
- Multiplication par un élément externe (⚠️ ne multiplie pas les éléments de l'ensemble avec eux mêmes, mais multiplie des éléments de $E$ par un scalaire $\in K$)

La loi interne $(E, +)$ est un groupe abelien/commutatif 
- L'addition $+$ est une LCI $\forall (x, y) \in E^2, \; x + y \in E$ (on dit que l'ensemble est stable pour la loi $+$)
- Tous éléments de $E$ a un symétrique, qui lui aussi appartient à l'ensemble 
- L'ensemble possède un élément central, neutre $0_E$

La loi externe $(E, *)$, une application qui de $K × E$ dans $E$ vérifiant:
- ∀(α, β) ∈ $K^2$ , ∀x ∈ E, (α + β)⋅x = α⋅x + β⋅x
- ∀α ∈ K, ∀(x, y) ∈ $E^2$, α⋅(x + y) = α⋅x + α⋅y
- ∀(α, β) ∈ $K^2$, ∀x ∈ E, α⋅(β⋅x) = (αβ)⋅x
- ∀x ∈ E, 1⋅x = x

On parle de **K-espace vectoriel** ou **espace vectoriel sur K**.


### Montrer qu'un espace est un espace vectoriel:

1. On se ramène un ensemble de référence classique connu A
1. Puis, montrer que l'ensemble E est un sous espace vectoriel de cet ensemble A

### 1. Methode 1:
<u>$F$ est un sous sev de $E$:</u>

1. $F \subset E$
2. $F \neq \emptyset$ (car $0_E \in F$)
3. $\forall (x, y) \in F^2, \; (x +y) \in F$  <!> Il se peut que la somme soit dans $E$ mais pas dans $F$.

4. $\forall \alpha \in F, \; \forall \lambda \in K, \; (\lambda*x) \in F$

On peut résumer les 2 priopriétées par:
$\forall (x, y) \in F^2, \forall (\lambda, \mu) \in K^2, \; (\lambda * x + \mu * y) \in F$


$\begin{cases} \text{Si $\lambda = 1$ et $\mu = 1$ on retombe sur 3.} \\ 
               \text{Si $y=0$ car $0_F$ et $\lambda = 1 \in F$, on retombe sur 4.} 
\end{cases}$

**Les autres propriétés telles que: l'associativité, commutativité... ne sont pas vérifiées pour F, car elles le sont déja pour E. Comme F est inclue dans E, tout ce qui est vrai pour E l'est aussi pour F.**

<span style="color:green"><u>Exemple:</u></span> Montrer que F est un sous espace vectoriel d'un espace plus grand.

$F =  {{x \choose y} \in \mathbb{R^2} \;\; / \;\; x + y = 0}$

![](Images/F_esv_R2.png)

1. $\mathbb{R^2}$ est un espace vectoriel, $F \subset \mathbb{R^2}$
1. Si ${x \choose y} = {0 \choose 0}$, alors: x + y = 0 + 0 = 0 $\implies 0_R^2 \subset F$
1. Montrer que peu importe les combinaisons linéaires qu'on fait: on se ramène à 0, car $x + y = 0$

Soient ${x \choose y} \in \mathbb{R^2}$, ${x^\prime \choose y^\prime} \in \mathbb{R^2}$ et $\lambda \in \mathbb{R}$

${x \choose y} + \lambda * {x^\prime \choose y^\prime} = 
 {x + \lambda * x^\prime \choose y + \lambda * y^\prime} = {0 \choose 0}$

$\implies (x + \lambda * x^\prime) + (y + \lambda * y^\prime) = 0$

$\implies (x + y) + \lambda *(x^\prime + y^\prime) = 0$

Comme $x + y = 0$ car $x + y \in \mathbb{R^2}$ et $x^\prime + y^\prime = 0$ car $x^\prime + y^\prime \in \mathbb{R^2}$

$\implies 0 + \lambda * 0 = 0$

$\implies F$ est un espace vectoriel.

<span style="color:blue">Application linéaire</span>
===

**Motivation**:

Beaucoup de problème se ramène à des systèmes d'équation linéaire. Eg. Représentation matricielle: $A.X = Y$

Cette représentation peut être réécrite par $U(x) = v$, où :
- U est une application linéaire
- $x$ est un vecteur d'un sous espace vectoriel fini/infini (ensemble des solutions) dans $E$ 
- $v$ est l'image de l'application linéaire $U$ d'autre espace $F$


Aussi appelée _opérateur linéaire_ ou _transformation linéaire_ est une application entre 2 espaces vectoriels sur un corps qui respecte **l'addition des vecteurs** et la **multiplication scalaire**, et préserve ainsi plus généralement les combinaisons linéaires.

Soient $\mathbb{E}$ et $\mathbb{F}$ deux espaces vectoriels sur un coprs $\mathbb{K}$

<span style="color:red">$f$ est une application linéaire de $E$ dans $F$ $ssi$</span>:

<span style="color:red">1. La loi $+ \; f(u + v) = f(u) + f(v), \; \forall u, v \in E$</span>

<span style="color:red">2. La loi $* \; f(\lambda . u) = \lambda . f(u), \; \forall u \in E, \; \forall \lambda \in K$</span>



* Une application linéaire de E dans F est aussi appelé <span style="color:yellow">**Morphisme ou Homo-Morphisme**</span>
    - L'ensemble des applications linéaires de $E$ dans $F$ est noté $L(E, F)$
* Une application linénaire de $E$ dans $E$ est appelé <span style="color:yellow">**Endormorphisme**</span> de $E$
    - L'ensemble des endormorphismes de $E$ est noté $L(E)$

<span style="color:green"><u>Cas particuliers</u></span>:

- Lorsque $f$ est **bijective**, $f$ est un **isomorphisme** de $E$ dans $F$.
- Lorsque $E =  F$, $f$ est un **endomorphisme**.
- Lorsque $E = F$ et $f$ est **bijective**, $f$ est un **automorphisme** de E.
- Lorsque $F = K$, $f$ est une forme linéaire.

<span style="color:green"><u>Exemple 1:</u></span>

$$f: R_3 \to R_2$$

$$(x, y, z) \to (-2x, y + 3z)$$

$f$ est une application linéaire:

* Soient $u = (x, y, z) \in R_3$ et $v = (x_1, y_1, z_1) \in R_3$

$f(u + v) = f(x + x_1, y + y_1, z + z_1)$

$f(u + v) = -2(x + x_1), (y + y_1) + 3(z + z_1)$

$f(u + v) = (-2x, y+ 3z) + (-2x_1 + y_1 + 3z_1)$

$f(u + v) = f(u) + f(v)$

* Soient $u = (x, y, z) \in R_3$ et $\lambda \in R_3$

$f(\lambda.u) = f(\lambda.x, \lambda.y, \lambda.z)$

$f(u + v) = \lambda * -2x, \lambda.y +\lambda.3z$

$f(u + v) = \lambda.(-2x, y+ 3z)$

$f(u + v) = \lambda.f(u)$

# <span style="color:blue"><u>Noyau</u></span>

Le Noyau, noté $Ker(\phi)$ est l'ensemble des éléments de $E$ qui ont pour image $0$.
$$Ker(\phi) = { x \in E, \; \phi(x) = 0}$$

- Les antécédants de $0_F$
- expo n'est pas une application linéaire $\phi(0_E) = 1 \neq 0_F$
- $\phi$ est injective ssi $\ker(\phi) = {0}$ c.a.d. le noyeu est réduit au vecteur nul
- Thèorème: $Ker(u)$ est un espace vectoriel.

<span style="color:green"><u>Démonstration:</u></span>

Supposons $Ker (\phi) = {0}$

Soit $(u_1, u_2) \in E^2$ et $\phi$ est une application linéaire.

$\phi(u_1) = \phi(u_2)$

$\iff \phi(u_1) - \phi(u_2) = 0$

$\iff \phi(u_1 - u_2) = 0$ par linéarité de $\phi$ 

$\iff u_1 - u_2 \in Ker(\phi)$ 

$\iff u_1 - u_2 = 0$ car $Ker (\phi) = {0}$, par supposition

$\iff u_1 = u_2)$ alors $\phi$ est injective



<span style="color:green"><u>Exemple 1:</u></span> Déterminer le noyau de cette application linéaire


$$f: R^2 \to R^2$$

$$(x, y) \to (x + y, x + y)$$ 

1. Déterminer une base:
Soit $v = (x, y) \in R^2$

Par définition, si $v \in Ker(f) \iff f(v) = 0_(R^2)$ (être dans le noyau, c'est être un antécédant de $R^2$)

$f(v) = 0_(R^2)$

$\iff (x + y, x + y) = (0, 0)$


$\iff \begin{cases} \text{x + y = 0} \\ 
               \text{x + y = 0}
\end{cases}$

$\iff \begin{cases} \text{x + y = 0} \end{cases}$

$\iff \begin{cases} \text{y = -x} \end{cases}$

$\iff v = (x, -x)$

$\iff v = x (1, -1)$, factorisation par x 

$\iff v \in Vect{(1, -1)}$, $v$ appartient à un certain sous expace engéndré par le vecteur $(1, -1)$

$\iff Ker(f) = Vect{(1, -1)} $, le $ker(f)$ est engendré par un vecteur non nul.

$\iff Dim(Ker(f)) = 1$

**Matrice d'ordre $n$**: Une matrice carré de dim $n$, notée $M_n(K)$ au lieu de $M_{(n, n)}(K)$.

**Matrice d'ordre $(m, n)$**: Une matrice $(m × n)$, notée $M_{(m, n)}(K)$, d’ordre $(m, n)$ ou de dimension $m × n$. 

**Matrice symétrique**: Une matrice carré qui est égale à sa propre transposée $\forall i, j \in [1; n] \; a_{i,j} = a_{j,i}$ où les $a_{i,j}$ sont les coefficients de la matrice et $n$ est son ordre.

**Matrice symétrique réelle définie positive**: Une matrice positive et inversible, autrement dit, elle vérifie l'une des 4 propriétés équivalentes suivantes : 

1.	Pour toute matrice colonne $x \neq 0$ à $n$ éléments réels, on a: $x^T M * x > 0$.
    Autrement dit, la forme *quadratique* définie par $M$ est strictement positive pour $x \neq 0$.

2.	Toutes les valeurs propres de $M$ sont strictement positives, c'est-à-dire : 
$SP(M) \in ]0, \infty]$

3.	La forme bilinéaire symétrique définie par la relation: $R^n * R^n \to R, \; <x, y> = x^T * M * y$, est un produit scalaire sur $R^n$

4. Il existe $N \in M_n(R)$ inversible telle que $M = N^T* N$ (autrement dit : $M$ est congruente à la matrice identité).

**Matrice symétrique réelle négative**: Si son opposée (symétrique elle aussi) est définie positive.

$$A*u - \lambda * u = 0$$

## régression logistique 

La régression logistique est un modèle de *classification* très utilisé en statistiques et notamment en apprentissage automatique. 

La régression logistique peut s’appliquer à la classification Binaire et à la classification Multi-classe.


#### Le concept de relation linéaire
“Fonction affine”, telle que $f(x)=ax+b$ avec $a$ le coefficient directeur et $b$ l’ordonnée à l’origine.

y = β₀ + β₁X₁ + β₂X₂ + … + β₈X₈

Xᵢ : iᵉᵐᵉ variable explicative "une colonne".

βᵢ : iᵉᵐᵉ coefficient directeur de l’hyper-droite associé à la iᵉᵐᵉ variable explicative et β₀ l’ordonnée à l’origine. Ici, on peut interpréter βᵢ comme une mesure de l’importance donnée à Xᵢ dans la classification : plus ce coefficient est élevé, plus Xᵢ joue un rôle dans l’output du modèle.

#### La fonction sigmoïde comme estimateur d’une probabilité

Une probabilité est un nombre entre 0 et 1, une idée importante en classification est qu’on cherche à estimer la probabilité d’appartenance à chaque classe pour ainsi être en mesure de classifier chaque nouvelle observation dans la classe associée à la probabilité la plus forte.
L’idée principale de la régression logistique est de se servir de la fonction sigmoïde pour transformer le nombre que donne l’hyper-droite en une probabilité et on définie un seuil e.g. 0.5.
